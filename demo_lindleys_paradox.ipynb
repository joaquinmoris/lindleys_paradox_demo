The [Lindley's Paradox](https:////en.wikipedia.org/wiki/Lindley%27s_paradox) was originally described originally in the 1930s by Harold Jeffrey, and discussed again by Dennis Lindley in 1957. Following the NHST paradigm, if we obtain a p-value that is less than our α criterion, we would reject the null hypothesis (H0) in favour of our alternative hypothesis (H1). However, this paradox shows that under some conditions, obtaining a p-value that is lower but close to our α criterion, is more likely in the case that H0 were true, than if H1 were true.

In this script we will show that if an experiment has a high statistical power, the Lindley's paradox appears as described.

## Parameters

In this section we will define the parameters of the simulations that we will run.
Two normal distributions will be randomly generated, with the mean and standard deviations defined below. Then, an independent samples t test will be calculated, and its corresponding p-value stored. We will do this many times, so that we can see the distribution of the p-values, when H0 is true (μ1 = μ2), and when H1 is true (μ1 ≠ μ2).
The standard deviation will be kept constant across groups and conditions to simplify the interpretation of the results.

```python
simulations = 10000   # How many repetitions of the t-test do we want to run?
sample_size = 1000    # What should the sample size per group be?
mean_group1 = 0       # Mean of group 1
mean_group2_H0 = 0    # Mean of group 2 (H0 condition)
mean_group2_H1 = 0.2  # Mean of group 2 (H1 condition)
sd_val = 1            # Standard deviation
```
